```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
file = os.path.join('data', 'AwardsPlayers.csv')
awards = pd.read_csv(file)
awards.head()
```

```python
file = os.path.join('data', 'People.csv')
people = pd.read_csv(file)
people.head()
```

```python
merge_table = pd.merge(people, awards, on="playerID", how="right")
merge_table
```

```python
list(merge_table.columns.values)
```

```python
merge_table.drop(['deathYear', 'deathMonth', 'deathDay', 'deathCountry', 'deathState', 'deathCity', 
                            'weight', 'height', 'bbrefID', 'debut',"birthMonth", "birthDay", "birthCountry", 
                 "birthState", "birthCity", "nameFirst", "nameLast", "nameGiven", "bats", "throws", "finalGame",
                 "retroID", "lgID"], axis = 1, inplace = True)
merge_table
```

```python
modern_df = merge_table.loc[merge_table['yearID'] > 1899.0]
modern_df
```

```python
modern_df['Age'] = modern_df['yearID'] - modern_df['birthYear']
modern_df
```

```python
modern_df["Average age of winner"] = modern_df.groupby("awardID")["Age"].transform('mean')
# hosted_in_us["pledged"] = hosted_in_us["pledged"].map("${:,.2f}".format)
modern_df["Average age of winner"] = modern_df["Average age of winner"].map("{:,.2f}".format).astype('float64') 
modern_df["birthYear"] = modern_df["birthYear"].map("{:.0f}".format).astype('float64') 
modern_df["Age"] = modern_df["Age"].map("{:.0f}".format).astype('float64') 
# cols = ['Age','birthYear']
# modern_df[cols] = modern_df[cols].replace(',','.', regex=True).astype(float)
modern_df
```
