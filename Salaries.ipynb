```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt 
import numpy
import seaborn as sns
```

```python
file = os.path.join('data', 'Salaries.csv')
salaries = pd.read_csv(file)
salaries.head()
```

```python
file = os.path.join('data', 'People.csv')
people = pd.read_csv(file)
people.head()
```

```python
merge_table = pd.merge(people, salaries, on="playerID", how="outer")
merge_table
```

```python
file = os.path.join('data', 'Pitching.csv')
pitching = pd.read_csv(file)
pitching.drop(columns=['stint','lgID', 'IBB','SV', 'WP','BK','BFP','GF', 'SH','SF'], inplace=True)
pitching.head()
```

```python
pitching["position"] = 'P'
pitching.head()
```

```python
list(pitching)
```

```python
pitching.drop(columns = [ 'yearID','teamID','W','L','G','GS','CG','SHO','IPouts','H','ER','HR','BB','SO','BAOpp','ERA','HBP',
               'R','GIDP',], inplace = True)
pitching.head()
```

```python
# df.drop(columns=['B', 'C'])
merge_table.drop(['deathYear', 'deathMonth', 'deathDay', 'deathCountry', 'deathState', 'deathCity', 
                            'weight', 'height', 'bbrefID', 'debut',"birthMonth", "birthDay", "birthCountry", 
                 "birthState", "birthCity", "nameFirst", "nameLast", "nameGiven", "bats", "throws", "finalGame",
                 "retroID", "teamID", "lgID"]
                 , axis = 1, inplace = True)
merge_table.fillna(0)
```

```python
new_age_table = merge_table.loc[merge_table['yearID'] > 1800.0]
new_age_table
```

```python
new_age_table["Age"] = new_age_table["yearID"] - new_age_table["birthYear"]
new_age_table

# new_age_table.style.format({'salary' : '{:+.2f}'})
```

```python
# df['freq'] = df.groupby('a')['a'].transform('count')
new_age_table['Seasons'] = new_age_table.groupby('playerID')['playerID'].transform('count')
new_age_table
```

```python
clean_df = new_age_table.loc[new_age_table['Seasons'] > 5]
clean_df
```

```python
under_df = clean_df.loc[new_age_table['Age'] < 46]
```

```python
plt.scatter(under_df["Age"], under_df["salary"], marker="o")
plt.title("Age vs Salary")
plt.ylabel("Salary")
plt.xlabel("Age")
plt.grid(True)
plt.ylim(0,35000000)
plt.savefig("Salary/Age vs Salary.png")
plt.show()
```

```python
average_salary = clean_df.groupby(["Age"])["salary"].mean()
print(average_salary)
```

```python
pd.set_option('float_format', '{:f}'.format)
```

```python
average_salary
```

```python
count_salary = clean_df.groupby(["Age"])["salary"].count()
count_salary
```

```python
pitching_salaries = pd.merge(under_df, pitching, on="playerID", how="outer")
pitching_salaries
```

```python
p_salaries = pitching_salaries.dropna()
p_salaries
```

```python
pitch = p_salaries.drop_duplicates().reset_index()
pitch
```

```python
P_df["Average Salary per Age for P"] = pitch.groupby("Age")["salary"].transform('mean')
P_df
```
