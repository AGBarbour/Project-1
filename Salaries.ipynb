```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt 
import numpy
import seaborn as sns
```

```python
file = os.path.join('data', 'Salaries.csv')
salaries = pd.read_csv(file)
salaries.head()
```

```python
file = os.path.join('data', 'People.csv')
people = pd.read_csv(file)
people.head()
```

```python
merge_table = pd.merge(people, salaries, on="playerID", how="outer")
merge_table
```

```python
unique_df = merge_table["playerID"].value_counts()
unique_df
```

```python
# df.drop(columns=['B', 'C'])
merge_table.drop(['deathYear', 'deathMonth', 'deathDay', 'deathCountry', 'deathState', 'deathCity', 
                            'weight', 'height', 'bbrefID', 'debut',"birthMonth", "birthDay", "birthCountry", 
                 "birthState", "birthCity", "nameFirst", "nameLast", "nameGiven", "bats", "throws", "finalGame",
                 "retroID", "teamID", "lgID"]
                 , axis = 1, inplace = True)
merge_table.fillna(0)
```

```python
new_age_table = merge_table.loc[merge_table['yearID'] > 1800.0]
new_age_table
```

```python
new_age_table["Age"] = new_age_table["yearID"] - new_age_table["birthYear"]
new_age_table

# new_age_table.style.format({'salary' : '{:+.2f}'})
```

```python
# df['freq'] = df.groupby('a')['a'].transform('count')
new_age_table['Seasons'] = new_age_table.groupby('playerID')['playerID'].transform('count')
new_age_table
```

```python
clean_df = new_age_table.loc[new_age_table['Seasons'] > 5]
clean_df
```

```python
clean_df = new_age_table.loc[new_age_table['Age'] < 46]
```

```python
plt.scatter(clean_df["Age"], clean_df["salary"], marker="o")
plt.title("Age vs Salary")
plt.ylabel("Salary")
plt.xlabel("Age")
plt.grid(True)
plt.ylim(0,35000000)
plt.savefig("Salary/Age vs Salary.png")
plt.show()
```

```python
average_salary = clean_df.groupby(["Age"])["salary"].mean()
print(average_salary)
```

```python
pd.set_option('float_format', '{:f}'.format)
```

```python
count_salary = clean_df.groupby(["Age"])["salary"].count()
count_salary
```

```python
# df['freq'] = df.groupby('a')['a'].transform('count')
clean_df["Average Salary per Age"] = clean_df.groupby("Age")["salary"].transform('mean')
clean_df
```

```python
plt.bar(clean_df['Age'], clean_df['Average Salary per Age'], color='r', alpha=0.5, align="center")
plt.title("Average Salary Per Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.savefig("Salary/Average Salary Per Age.png")
plt.show()
```

```python
sns.regplot(x='Age', y='salary', data=clean_df, order = 2)
plt.ylim(0,35000000)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.swarmplot(x='Age', y='salary', data=clean_df)
plt.xticks(rotation='vertical')
plt.ylim(0,35000000)
plt.title("Salary VS Age")
plt.xlabel("Age")
plt.ylabel("Salary (in tens of millions)")
plt.show()
```

```python
sns.set_style('darkgrid')
sns.boxplot(x='Age', y='salary', data=clean_df, palette='Blues')
plt.xticks(rotation='vertical')
plt.ylim(0,35000000)
plt.show()
```
